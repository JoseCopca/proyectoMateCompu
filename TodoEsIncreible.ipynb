Se programaron cuatro clases, una llamada Cell que contiene el estado de una celda del autómata celular y las diferentes reglas que lo modificarán en los siguientes pasos. De este modo se pudo asegurar la sincronía a la hora de modificar el estado del bosque o la pila de arena.

```python
class Cell:
    def __init__(self,state,edge=False):
        """
        parametros:
            state -> el estado, ya sea la altura de la arena, o si el arbol esta vivo, muerto o quemandose
            edge -> solo sirve para la arena (en caso de que este en una caja, para que no se pueda "caer")
        """
        self.state=state
        self.edge=edge
        
    """
    estadoActual es la vecindad (ya sea de 8 o de 2), ya que de esta depende lo que le suceda al estado de la celda
    """
        
    #todo esto es de arena 1d            
    def nuevoEstadoCaja(self,estadoActual,sc):
        """
        no se puede caer por los bordes, solo se acumula en las orillas
        """
        if estadoActual[0]-self.state>sc: #h_i-1 - h_i
            self.state+=1
        if not self.edge:
            if self.state-estadoActual[2]>sc:#h_i - h_i+1
                self.state-=1
        
    
    def nuevoEstadoMesa(self,estadoActual,sc):
        """
        la arena puede desaparecer si se trata de una orilla
        """
        if self.state-estadoActual[2]>sc: #h_i - h_i+1
            self.state-=1
        if estadoActual[0]-self.state>sc: #h_i-1 - h_i
            self.state+=1

    def agrega(self):
        self.state+=1
    
    def random(self):
        self.state=randint(0,5)
    
    def wipe(self):
        self.state=0
        
    #esto es bosque
    def nuevoEstadoBosque(self, estadoActual, f,p):
        """
        revisa su estado y el estado de los 8 a su alrededor para decidir su proximo estado
        """
        if self.state==1:
            self.checaFuego(estadoActual)
            #if self.state==1:#este if es por si el estado sigue siendo vivo (despues de checar el fuego de los vecinos)
            self.spontaneousCombustion(f)
        else:
            self.muere()
            self.revive(p)
    
    def checaFuego(self, estadoActual):
        """
        revisa que los de alrededor no se esten quemando
        """
        for i in xrange(0,3,1):
            for j in xrange(0,3,1):
                if estadoActual[i,j]==2:
                    self.state=2
    
    def spontaneousCombustion(self,f):
        """
        tiene una probabilidad para arder en llamas repentinamente
        """
        if self.state==1 and random()<=f:
            self.state = 2
    
    def muere(self):
        """
        si el estado es en llamas, ahora esta muerto
        """
        if self.state==2:
            self.state = 0
    
    def revive(self,p):
        """
        hay una probabilidad de que el arbol reviva
        """
        if self.state == 0 and random()<=p: #por que no random.random()??
            self.state=1
            
    def algunasEnLlamas(self):
        if random()<0.30:
            self.state=2
        else:
            self.state=1
            
    #esto es pila de arena2
    def nuevoEstadoArena2(self,estadoActual,K):
        """
        checa no ser mayor a K, o que los 4 vecinos no sean mayores a K
        """
        if self.state>K:
            self.state-=4
        if estadoActual[0,1]>K:
            self.state+=1
        if estadoActual[1,0]>K:
            self.state+=1
        if estadoActual[1,2]>K:
            self.state+=1
        if estadoActual[2,1]>K:
            self.state+=1
    
    """
    las demas funciones de pila de arena 2 son iguales a las de pila de arena 1 y no tenia caso copiarlas dos veces
    """
```

Las tres clases faltantes son las siguientes. Es una clase por cada proyecto. Primero se crea una matriz vacía con las dimensiones adecuadas al modelo. Después, cada celda tiene una dirección de memoria de la clase cell. La función $dinamica$ va a cambiar el estado de cada celda (utilizando siempre la clase cell) y como es un manejo de objetos al cambiar la celda se cambia la matriz.

```python
class PilaArena:
    def __init__(self,L):
        """
        parametros:
        L -> longitud del arreglo
        sc -> pendiente critica
        estadoActual -> un arreglo de tamanio L de Cells
        """
        self.L=L
        self.sc=2
        self.estadoActual=np.empty((1,self.L), dtype=object)
        
    def crea(self):
        """
        llena el arreglo con Cells (cuyo estado es 0)
        """
        for i in xrange(0,self.L,1):
            self.estadoActual[0,i]=Cell(0)
        
    
    def dinamicaMesa(self):
        """
        revisa que la pila sea estable, si lo es, agrega aleatoriamente un grano de arena, 
        si no lo es, cambia el estado de todas las Cells, en base al estado que ya se tiene (para eso "estatico")
        """
        estatico=self.creaEstadoEstatico()
        estable=True
        for i in xrange(1,self.L-1,1):
            if estatico[0,i]-estatico[0,i+1]>self.sc:
                estable=False
        if estable:
            self.estadoActual[0,randint(1,self.L-1)].agrega()
        else:
            for i in xrange(1,self.L-1,1):
                self.estadoActual[0,i].nuevoEstadoMesa(estatico[0,i-1:i+2],self.sc)
    
    def dinamicaCaja(self):
        """
        revisa que la pila sea estable, si lo es, agrega aleatoriamente un grano de arena, 
        si no lo es, cambia el estado de todas las Cells, en base al estado que ya se tiene (para eso "estatico")
        """
        estatico=self.creaEstadoEstatico()
        estable=True
        for i in xrange(1,self.L-1,1):
            if estatico[0,i]-estatico[0,i+1]>self.sc:
                estable=False
        if estable:
            self.estadoActual[0,randint(1,self.L-1)].agrega()
        else:
            for i in xrange(1,self.L-1,1):
                self.estadoActual[0,i].nuevoEstadoCaja(estatico[0,i-1:i+2],self.sc)
        
        
    def creaEstadoEstatico(self):
        """
        crea un arreglo con los estados de las Cells (no las Cells en si, ya que solo son necesarios los nombres) 
        """
        estatico=np.empty((1,self.L))
        i=0
        for celda in self.estadoActual[0,:]:
            estatico[0,i]=celda.state
            i+=1
        return estatico
    
    def toString(self):
        """
        crea un arreglo con los estados de las Cells para imprimirlo
        ya que imprimir un arreglo de objetos te da sus direcciones en memoria
        """
        imprime=np.empty((1,self.L))
        for i in xrange(0,self.L,1):
            imprime[0,i]=self.estadoActual[0,i].state
        print imprime[0,1:self.L-1]
        #print self.estadoActual
    
    def startRandom(self):
        """
        se asigna valor aleatorio a las Cells 
        """
        for i in xrange(1,self.L-1,1):
            self.estadoActual[0,i].random()
            
    def wipe(self):
        """
        borra las orillas (que de hecho nunca se despliegan)
        esto permite que la pila de arena se desparrame como si estuviera sobre una mesa
        """
        self.estadoActual[0,0].wipe()
        self.estadoActual[0,self.L-1].wipe()
        
    def loopMesa(self, steps):
        """
        Ejecuta el numero especificado de pasos
        """
        [self.stepMesa() for i in xrange(steps)]
    
    def loopCaja(self, steps):
        """
        Ejecuta el numero especificado de pasos
        """
        [self.stepCaja() for i in xrange(steps)]

    def stepMesa(self):
        """
        Avanza un paso t -> t+1
        """
        self.dinamicaMesa()
        self.wipe()
    
    def stepCaja(self):
        """
        Avanza un paso t -> t+1
        """
        self.dinamicaCaja()
        self.wipe()
```

```python
class Bosque():
    def __init__(self,L,f,p):
        """
        parametros:
        L -> longitud de los lados del arreglo (el arreglo tiene LxL celdas)
        p -> probabilidad de que un arbol reviva
        f -> probabilidad de que un arbol arda en llamas espontaneamente
        estadoActual -> un arreglo de tamanio L vacio de tipo object para llenarlo de Cells

        estadoEnElTiempo -> lista donde guardas numero de vivos, muertos y en llamas
        """
        self.L=L
        self.estadoActual=np.empty((self.L,self.L), dtype=object)
        self.f=f
        self.p=p

        self.estadoEnElTiempo=[]
        
        
    def __getitem__(self):
        return np.array(estadoEnElTiempo)
        
    def crea(self):
        """
        llena todo el arreglo de Cells con estado 1 (excepto el marco)
        """
        for i in xrange(1,self.L-1,1):
            for j in xrange(1,self.L-1,1):
                self.estadoActual[i,j]=Cell(1)
        self.marco()
        
    def marco(self):
        """
        crea un marco de 42 alrededor del bosque
        esto es para no tener problemas con las orillas
        """
        self.estadoActual[:,0]=Cell(42)
        self.estadoActual[0,:]=Cell(42)
        self.estadoActual[self.L-1,:]=Cell(42)
        self.estadoActual[:,self.L-1]=Cell(42)
        #because life the universe and everything
        
    
    def dinamica(self):
        """
        cambia el estado de todas las Cells basados en el estado actual (que se guarda en estatico)
        """
        estatico=self.creaEstadoEstatico()
        for i in xrange(1,self.L-1,1):
            for j in xrange(1,self.L-1,1):
                pedazo=estatico[i-1:i+2,j-1:j+2]
                self.estadoActual[i,j].nuevoEstadoBosque(pedazo,self.f,self.p)
    
    def empiezaFuego(self):
        """
        funcion semi-provisional para poder probar el bosque con un "solo germen"
        """
        self.estadoActual[2,2]=Cell(2)
                
        
    def creaEstadoEstatico(self):
        """
        vacia todos los estados de los Cell de estado actual en estatico (para que no se mueva porque los objetos funcionan por referencia)
        """
        estatico=np.empty((self.L,self.L))
        for i in xrange(0,self.L,1):
            for j in xrange(0,self.L,1):
                estatico[i,j]=self.estadoActual[i,j].state
        return estatico
    
    def toString(self):
        """
        crea un arreglo con los estados de las Cells para imprimirlo
        ya que imprimir un arreglo de objetos te da sus direcciones en memoria
        """
        imprime=np.empty((self.L,self.L))
        for i in xrange(0,self.L,1):
            for j in xrange(0,self.L,1):
                imprime[i,j]=self.estadoActual[i,j].state
        return imprime[1:self.L-1,1:self.L-1]
        
    def cuentaEstados(self):
        vivos=0
        muerto=0
        enllamas=0
        for i in xrange(0,self.L,1):
            for j in xrange(0,self.L,1):
                if self.estadoActual[i,j].state==1:
                    vivos+=1
                elif self.estadoActual[i,j].state==2:
                    enllamas+=1
                elif self.estadoActual[i,j].state==0:
                    muerto+=1
        self.estadoEnElTiempo.append((vivos,enllamas,muerto))
        
    def loop(self, steps=10):
        """
        Ejecuta el numero especificado de pasos
        """
        [self.step() for i in xrange(steps)]

    def step(self):
        """
        Avanza un paso t -> t+1
        """
        self.dinamica()
        self.cuentaEstados()
        self.imprime()
    
    def startFuego(self):
        """
        se asigna valor a las Cells (vivas o quemandose)
        """
        for i in xrange(1,self.L-1,1):
            for j in xrange(1,self.L-1,1):
                self.estadoActual[i,j].algunasEnLlamas()
    
    def imprime(self):
        """
        muestra los diferentes numeros de diferentes colores
        pero solo lo hace con el ultimo
        """

        estatico=self.creaEstadoEstatico()
        colores=plt.imshow(estatico[1:self.L-1,1:self.L-1])
        print colores
```
